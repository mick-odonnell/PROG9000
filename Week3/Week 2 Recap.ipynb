```python
# 3.  Write a program that makes a shopping list, with the user able to enter an item and the number of items required. 
# Then print the list, and the total number of items needed. Hint below if needed.

s_list_items = []
s_list_ct = []

make_list = True

while make_list:
    shopping_item = input("Please type the item you want to add to the list")
    shopping_item_count = int(input("Please add the count of " + shopping_item + "(s) that you require"))

    s_list_items.append(shopping_item)
    s_list_ct.append(shopping_item_count)
    
    choice = input("Do you wish to add another item to the shopping list: Y or N?")
    if choice == 'N':
        make_list = False
    if choice != 'Y':
        print("Your choice in invalid")
        break
        
for i in range(len(s_list_items)):
    print(s_list_items[i], ":     ", s_list_ct[i])

    
        
```

```python
# 4. Write a function which finds the largest of a list of numbers. 
# Do not use a method, compare each element to the current maximum. Use a loop.

nums = [3, 6, 8, 28, 1, -34, 17]

top = nums[0]

for i in range(len(nums) - 1):
    comp = nums[i + 1]
    
    if comp > top:
        top = comp

print("The largest number is", top)
```

```python
# 5. Create a program that gives the user a menu with the following options:
#     - A Add an element to a list
#     - R Remove an element from the list
#     - P Replace an element with another
#     - S sort the list
#     Original list: [‘Madrid’, ‘Dublin’, ’New York’, ’Paris’, ’London’]

a_list = ['Madrid', 'Dublin', 'New York', 'Paris', 'London']

choice = input("Hello, welcome to list-builder.\n" +
      "What would you like to do?\n\n" +
      "Add an element to the list: Type A\n" +
      "Remove an element from the list: Type R\n" +
      "Replace an element with another: Type P\n" +
      "Sort the list: Type S\n")

# start with the easiest - adding an element:
if choice == 'A':
    add_item = input("Please type what you'd like to add:")
    a_list.append(add_item)
elif choice == 'R':
    # create a printing string of the items:
    print_str = ""
    for i in a_list:
        print_str += i + ':'
    remove_item = input("Which item would you like to remove:\n" +
                        print_str + "\n" + "Please type your choice...")
    if remove_item in a_list:
        a_list.remove(remove_item)
        for i in a_list:
            print(i, '\n')
    else:
        print('You made an invalid choice')
elif choice == 'P':
    print_str = ""
    for i in a_list:
        print_str += i + ':'
    replace_item = input("Which item would you like to remove:\n" +
                        print_str + "\n" + "Please type your choice...")
    if remove_item in a_list:
        new_item = input("Please enter the new item")
        item_index = a_list.index(replace_item)
        a_list.insert(item_index, new_item)
        a_list.remove(replace_item)
        for i in a_list:
            print(i)
    else:
        print("You made an invalid choice")
elif choice == 'S':
    a_list.sort()
    for i in a_list:
        print(i)
else:
    print("You made an invalid choice")
                    
```
